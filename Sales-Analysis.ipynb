```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example: Line chart for sales over time
df['Order Date'] = pd.to_datetime(df['Order Date'])  # Ensure the date column is in datetime format
df.sort_values('Order Date', inplace=True)  # Sort by date

plt.figure(figsize=(10, 6))
plt.plot(df['Order Date'], df['Total Revenue'], label='Total Revenue', color='blue')
plt.title('Total Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example: Bar chart for average revenue by region
avg_revenue_by_region = df.groupby('Region')['Total Revenue'].mean()

plt.figure(figsize=(10, 6))
avg_revenue_by_region.plot(kind='bar', color='skyblue')
plt.title('Average Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Average Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Histogram of Total Revenue
plt.figure(figsize=(10, 6))
plt.hist(df['Total Revenue'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example: Scatter plot of Units Sold vs. Total Revenue
plt.figure(figsize=(10, 6))
plt.scatter(df['Units Sold'], df['Total Revenue'], alpha=0.7, color='purple')
plt.title('Units Sold vs. Total Revenue')
plt.xlabel('Units Sold')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```
